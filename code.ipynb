```python
columns_needed = [
    "country",
    "year",
    "population",
    "iso_code",
    "gdp",
    "biofuel_electricity",
    "coal_electricity",
    "electricity_demand",
    "electricity_generation",
    "fossil_electricity",
    "gas_electricity",
    "greenhouse_gas_emissions",
    "hydro_electricity",
    "nuclear_electricity",
    "oil_electricity",
    "renewables_electricity",
    "solar_electricity",
    "wind_electricity",
]

df = df[columns_needed]
df.head()
```

```python
# Find missing values
missing_values = df.isnull().sum()


for column in df.columns:
    print(f"{column}: {missing_values[column]/len(df)*100:.2f}")
```

```python
# 1. Number of years of data per country
years_per_country = df.groupby("country").size()

# 2. Count missing values per country for each variable
missing_counts = (
    df
    .groupby("country")
    .apply(lambda x: x.isna().sum())
)

# 3. Convert counts to percentages (normalized by years available)
missing_pct = missing_counts.div(years_per_country, axis=0) * 100

# 4. Inspect results
missing_pct.tail()
```

```python
df = df[df["year"] >= 1985].copy()

print("Dataset restricted to years >= 1985.")
print(f"Year range after filtering: {df['year'].min()} - {df['year'].max()}")
```

```python
# Columns that must meet the data-quality threshold
critical_cols = [
    "iso_code",
    "population",
    "electricity_generation",
    "electricity_demand"
]

# Kosovo is a real country but has no ISO code — exempt it explicitly
EXEMPT = {"Kosovo"}

# Identify countries that violate the 20% missing-data rule
countries_to_drop = missing_pct[
    (missing_pct[critical_cols] > 80).any(axis=1)
].index

countries_to_drop = [c for c in countries_to_drop if c not in EXEMPT]

# Drop those countries from the main dataframe
df_clean = df[~df["country"].isin(countries_to_drop)].copy()

print(f"Dropped {len(countries_to_drop)} countries due to excessive missing data.")
print(f"Remaining countries: {df_clean['country'].nunique()}")
print(f"(Kosovo retained as explicit exemption despite missing iso_code)")
```

solar and wind need to be handled differently from the other columns.
looking at the raw data, every pre-2000 non-null value for both columns is
already 0.0 - so the NaNs before 2000 aren't really missing, they just mean
the country had no capacity yet. same logic applies to post-2000 leading NaNs
(e.g. Bosnia had no wind farms until 2011, so those early 2000s rows should
also be zero, not interpolated). for 2023/2024 it's just a reporting lag so
we carry the last known value forward.

```python

renewable_cols = ["solar_electricity", "wind_electricity"]

def impute_solar_wind(df, col):
    result = df[col].copy()

    for country, group in df.groupby("country"):
        idx = group.index
        years = group["year"]
        vals = group[col]

        # anything before 2000 is a true zero
        pre2000 = idx[years < 2000]
        result.loc[pre2000] = result.loc[pre2000].fillna(0.0)

        post2000_idx = idx[years >= 2000]
        post2000_vals = vals.loc[post2000_idx]

        # if there's no post-2000 data at all just zero fill and move on
        if post2000_vals.isna().all():
            result.loc[post2000_idx] = 0.0
            continue

        first_valid = post2000_vals.first_valid_index()
        last_valid = post2000_vals.last_valid_index()

        # leading NaNs post-2000 = country hadn't built capacity yet, so zero
        leading = post2000_idx[post2000_idx < first_valid]
        result.loc[leading] = result.loc[leading].fillna(0.0)

        # trailing NaNs = 2023/2024 reporting lag, carry last known value
        trailing = post2000_idx[post2000_idx > last_valid]
        if len(trailing) > 0:
            result.loc[trailing] = result.loc[trailing].fillna(post2000_vals.loc[last_valid])

    return result


for col in renewable_cols:
    before = df_clean[col].isna().sum()
    df_clean[col] = impute_solar_wind(df_clean, col)
    after = df_clean[col].isna().sum()
    print(f"{col}: {before} missing -> {after} missing (filled {before - after})")


# quick checks
print()
for col in renewable_cols:
    print(f"{col} negatives: {(df_clean[col] < 0).sum()}")
    pre2000 = df_clean.loc[df_clean["year"] < 2000, col]
    print(f"{col} pre-2000 remaining NaN: {pre2000.isna().sum()}")

# spot check a few countries that had leading post-2000 NaNs
spot_check = {
    "Bosnia and Herzegovina": ("wind_electricity", 2000, 2010),
    "Malta": ("wind_electricity", 2000, 2011),
    "Martinique": ("solar_electricity", 2000, 2005),
}
print()
for country, (col, y1, y2) in spot_check.items():
    vals = df_clean[
        (df_clean["country"] == country) & df_clean["year"].between(y1, y2)
    ][["year", col]]
    print(f"{country} {col} {y1}-{y2}:")
    print(vals.to_string(index=False), "\n")

# check trailing fill worked
print("Uganda solar 2021-2024:")
print(df_clean[(df_clean["country"] == "Uganda") & (df_clean["year"] >= 2021)][["year", "solar_electricity"]].to_string(index=False))
```

```python
# ── Fossil: fossil_electricity = coal + oil + gas ─────────────────────────────
FOSSIL_COLS = ["coal_electricity", "oil_electricity", "gas_electricity", "fossil_electricity"]

print("Missing BEFORE fossil imputation:")
print(df_clean[FOSSIL_COLS].isnull().sum().to_string())

# Case A: gas missing, coal + oil + fossil present → gas ≈ 0
mask_a = (
    df_clean["coal_electricity"].notna()
    & df_clean["oil_electricity"].notna()
    & df_clean["gas_electricity"].isna()
    & df_clean["fossil_electricity"].notna()
)
df_clean.loc[mask_a, "gas_electricity"] = 0.0
print(f"\nCase A — gas_electricity = 0: {mask_a.sum()} rows")

# Case B: oil missing, coal + gas + fossil present → oil = fossil - coal - gas
mask_b = (
    df_clean["coal_electricity"].notna()
    & df_clean["oil_electricity"].isna()
    & df_clean["gas_electricity"].notna()
    & df_clean["fossil_electricity"].notna()
)
df_clean.loc[mask_b, "oil_electricity"] = (
    df_clean.loc[mask_b, "fossil_electricity"]
    - df_clean.loc[mask_b, "coal_electricity"]
    - df_clean.loc[mask_b, "gas_electricity"]
)
print(f"Case B — oil_electricity = fossil - coal - gas: {mask_b.sum()} rows")

# Case C: coal missing, oil + gas + fossil present → coal = fossil - oil - gas
mask_c = (
    df_clean["coal_electricity"].isna()
    & df_clean["oil_electricity"].notna()
    & df_clean["gas_electricity"].notna()
    & df_clean["fossil_electricity"].notna()
)
df_clean.loc[mask_c, "coal_electricity"] = (
    df_clean.loc[mask_c, "fossil_electricity"]
    - df_clean.loc[mask_c, "oil_electricity"]
    - df_clean.loc[mask_c, "gas_electricity"]
)
print(f"Case C — coal_electricity = fossil - oil - gas: {mask_c.sum()} rows")

# Case D: fossil missing, all 3 components present → fossil = coal + oil + gas
mask_d = (
    df_clean["coal_electricity"].notna()
    & df_clean["oil_electricity"].notna()
    & df_clean["gas_electricity"].notna()
    & df_clean["fossil_electricity"].isna()
)
df_clean.loc[mask_d, "fossil_electricity"] = (
    df_clean.loc[mask_d, "coal_electricity"]
    + df_clean.loc[mask_d, "oil_electricity"]
    + df_clean.loc[mask_d, "gas_electricity"]
)
print(f"Case D — fossil_electricity = coal + oil + gas: {mask_d.sum()} rows")

print("\nMissing AFTER fossil imputation:")
print(df_clean[FOSSIL_COLS].isnull().sum().to_string())

# Verification
check_f = df_clean[FOSSIL_COLS].dropna().copy()
check_f["computed"] = check_f[["coal_electricity", "oil_electricity", "gas_electricity"]].sum(axis=1)
check_f["diff"] = (check_f["computed"] - check_f["fossil_electricity"]).abs()
print(f"\nVerification — rows with all 4 columns: {len(check_f)}")
print(f"  Max |coal+oil+gas - fossil|: {check_f['diff'].max():.4f}")
print(f"  Rows with diff > 0.1:        {(check_f['diff'] > 0.1).sum()}")
print(f"  Negative oil values:         {(df_clean['oil_electricity'] < 0).sum()}")
print(f"  Negative gas values:         {(df_clean['gas_electricity'] < 0).sum()}")
print(f"  Negative coal values:        {(df_clean['coal_electricity'] < 0).sum()}")
```

```python
# ── Renewables: renewables_electricity = solar + hydro + wind + biofuel ───────
RENEW_COMPS = ["solar_electricity", "hydro_electricity", "wind_electricity", "biofuel_electricity"]
RENEW_TOTAL = "renewables_electricity"
THRESHOLD = 0.001  # TWh — prevents incorrect zero-fills on near-zero residuals

print("Missing BEFORE renewables imputation:")
print(df_clean[RENEW_COMPS + [RENEW_TOTAL]].isnull().sum().to_string())

cells_filled = 0

# Case A: exactly 1 component missing, others + renewables present
#         compute residual; fill if ≥ 0 (zero-fill if residual ≈ 0, value if positive)
for col in RENEW_COMPS:
    others = [c for c in RENEW_COMPS if c != col]
    mask = (
        df_clean[col].isna()
        & df_clean[others].notna().all(axis=1)
        & df_clean[RENEW_TOTAL].notna()
    )
    if mask.sum() == 0:
        continue
    residual = df_clean.loc[mask, RENEW_TOTAL] - df_clean.loc[mask, others].sum(axis=1)
    zero_idx = residual[residual.abs() < THRESHOLD].index
    pos_idx  = residual[residual >= THRESHOLD].index
    neg_idx  = residual[residual < -THRESHOLD].index
    df_clean.loc[zero_idx, col] = 0.0
    df_clean.loc[pos_idx,  col] = residual[pos_idx]
    n = len(zero_idx) + len(pos_idx)
    cells_filled += n
    print(f"Case A — {col}: {n} rows filled  ({len(neg_idx)} negative residuals skipped)")

# Case B: renewables total missing, all 4 components present → sum them
mask_total = (
    df_clean[RENEW_TOTAL].isna()
    & df_clean[RENEW_COMPS].notna().all(axis=1)
)
df_clean.loc[mask_total, RENEW_TOTAL] = df_clean.loc[mask_total, RENEW_COMPS].sum(axis=1)
cells_filled += mask_total.sum()
print(f"Case B — renewables_electricity = sum of 4 components: {mask_total.sum()} rows")

print(f"\nTotal cells filled: {cells_filled}")
print("\nMissing AFTER renewables imputation:")
print(df_clean[RENEW_COMPS + [RENEW_TOTAL]].isnull().sum().to_string())

# Verification
check_r = df_clean[RENEW_COMPS + [RENEW_TOTAL]].dropna().copy()
check_r["computed"] = check_r[RENEW_COMPS].sum(axis=1)
check_r["diff"] = (check_r["computed"] - check_r[RENEW_TOTAL]).abs()
print(f"\nVerification — rows with all 5 columns: {len(check_r)}")
print(f"  Max |sum_of_4 - renewables|: {check_r['diff'].max():.4f}")
print(f"  Rows with diff > 0.1:        {(check_r['diff'] > 0.1).sum()}")
```

```python
# ── Generation: electricity_generation = fossil + renewables + nuclear ─────────
GEN_PARTS = ["fossil_electricity", "renewables_electricity", "nuclear_electricity"]
GEN_TOTAL = "electricity_generation"

print("Missing BEFORE generation imputation:")
print(df_clean[[GEN_TOTAL] + GEN_PARTS].isnull().sum().to_string())

# Case A: generation missing, all 3 parts present → sum them
mask_gen = (
    df_clean[GEN_TOTAL].isna()
    & df_clean[GEN_PARTS].notna().all(axis=1)
)
df_clean.loc[mask_gen, GEN_TOTAL] = df_clean.loc[mask_gen, GEN_PARTS].sum(axis=1)
print(f"\nCase A — electricity_generation = fossil + renewables + nuclear: {mask_gen.sum()} rows")

# Case B: one part missing, others + generation present → solve by residual
for col in GEN_PARTS:
    others = [c for c in GEN_PARTS if c != col]
    mask = (
        df_clean[col].isna()
        & df_clean[others].notna().all(axis=1)
        & df_clean[GEN_TOTAL].notna()
    )
    if mask.sum() == 0:
        continue
    residual = df_clean.loc[mask, GEN_TOTAL] - df_clean.loc[mask, others].sum(axis=1)
    pos_idx = residual[residual >= 0].index
    neg_idx = residual[residual < 0].index
    df_clean.loc[pos_idx, col] = residual[pos_idx]
    print(f"Case B — {col} = generation - others: {len(pos_idx)} rows filled  ({len(neg_idx)} negative skipped)")

print("\nMissing AFTER generation imputation:")
print(df_clean[[GEN_TOTAL] + GEN_PARTS].isnull().sum().to_string())

# Verification
check_g = df_clean[[GEN_TOTAL] + GEN_PARTS].dropna().copy()
check_g["computed"] = check_g[GEN_PARTS].sum(axis=1)
check_g["diff"] = (check_g["computed"] - check_g[GEN_TOTAL]).abs()
print(f"\nVerification — rows with all 4 columns: {len(check_g)}")
print(f"  Max |fossil+renewables+nuclear - generation|: {check_g['diff'].max():.4f}")
print(f"  Rows with diff > 0.1: {(check_g['diff'] > 0.1).sum()}")
```

```python
# ── Forward-fill trailing NaNs (reporting lag) ────────────────────────────────
# Leading NaNs (country hadn't started reporting) are left as-is.
# Trailing NaNs (2023/2024 reporting lag) are filled with the last known value.
FFILL_COLS = [
    "gdp",
    "electricity_demand",
    "electricity_generation",
    "nuclear_electricity",
    "greenhouse_gas_emissions",
    "coal_electricity",
]

print("Missing BEFORE forward-fill:")
print(df_clean[FFILL_COLS].isnull().sum().to_string())

df_clean = df_clean.sort_values(["country", "year"]).reset_index(drop=True)

for col in FFILL_COLS:
    before = df_clean[col].isna().sum()
    df_clean[col] = df_clean.groupby("country")[col].transform(
        lambda s: s.ffill()
    )
    after = df_clean[col].isna().sum()
    print(f"  {col}: {before - after} trailing cells filled  ({after} remaining leading NaNs)")

print("\nMissing AFTER forward-fill:")
print(df_clean[FFILL_COLS].isnull().sum().to_string())
```

```python
# ── Zero-fill leading NaNs for capacity columns ───────────────────────────────
# No data before a country's first reported year = no capacity, not missing reporting.
# Only leading NaNs are zero-filled; trailing and interior NaNs are left as-is.
CAPACITY_COLS = ["coal_electricity", "nuclear_electricity", "hydro_electricity", "biofuel_electricity"]

print("Missing BEFORE capacity zero-fill:")
print(df_clean[CAPACITY_COLS].isnull().sum().to_string())

df_clean = df_clean.sort_values(["country", "year"]).reset_index(drop=True)

for col in CAPACITY_COLS:
    before = df_clean[col].isna().sum()
    def zero_fill_leading(s):
        first_valid = s.first_valid_index()
        if first_valid is None:
            return s.fillna(0.0)          # all NaN — country never reported, safe to zero
        s = s.copy()
        s.loc[:first_valid].fillna(0.0, inplace=True)
        return s
    df_clean[col] = df_clean.groupby("country")[col].transform(zero_fill_leading)
    after = df_clean[col].isna().sum()
    print(f"  {col}: {before - after} leading cells zero-filled  ({after} remaining)")

print("\nMissing AFTER capacity zero-fill:")
print(df_clean[CAPACITY_COLS].isnull().sum().to_string())
```

```python
# ── Backward trend extrapolation for leading NaNs (1985–reporting start) ──────
# For each country, fit a linear trend on the first available data points
# and extrapolate backward to fill years before reporting began.
import numpy as np

BACKCAST_COLS = [
    "gdp",
    "electricity_demand",
    "electricity_generation",
    "fossil_electricity",
    "renewables_electricity",
    "greenhouse_gas_emissions",
]

print("Missing BEFORE backward extrapolation:")
print(df_clean[BACKCAST_COLS].isnull().sum().to_string())

df_clean = df_clean.sort_values(["country", "year"]).reset_index(drop=True)

for col in BACKCAST_COLS:
    before = df_clean[col].isna().sum()

    def backcast(s, _col=col):
        if s.notna().sum() == 0:
            return s  # no data at all — leave as NaN

        first_valid_pos = s.first_valid_index()
        leading_idx = s.loc[:first_valid_pos].index[:-1]  # NaN indices before first valid

        if len(leading_idx) == 0:
            return s  # no leading NaNs for this country

        valid = s.dropna()

        if len(valid) < 2:
            # Not enough points for a trend — back-fill with constant
            s = s.copy()
            s.loc[leading_idx] = valid.iloc[0]
            return s

        # Fit linear trend on first 5 available data points
        anchor = valid.iloc[:5]
        years = df_clean.loc[anchor.index, "year"].values.astype(float)
        values = anchor.values.astype(float)
        slope, intercept = np.polyfit(years, values, 1)

        s = s.copy()
        for i in leading_idx:
            year = df_clean.loc[i, "year"]
            s.loc[i] = max(slope * year + intercept, 0.0)  # clamp at 0

        return s

    df_clean[col] = df_clean.groupby("country")[col].transform(backcast)
    after = df_clean[col].isna().sum()
    print(f"  {col}: {before - after} cells back-extrapolated  ({after} remaining)")

print("\nMissing AFTER backward extrapolation:")
print(df_clean[BACKCAST_COLS].isnull().sum().to_string())
```

```python
# ── GDP imputation via regional median GDP per capita ─────────────────────────
# Countries with no World Bank GDP data are imputed using the median GDP per
# capita of their region in each year, scaled by their own population.

REGION_MAP = {
    # Latin America & Caribbean
    "Antigua and Barbuda":               "Latin America & Caribbean",
    "Argentina":                         "Latin America & Caribbean",
    "Aruba":                             "Latin America & Caribbean",
    "Bahamas":                           "Latin America & Caribbean",
    "Belize":                            "Latin America & Caribbean",
    "Bermuda":                           "Latin America & Caribbean",
    "Bolivia":                           "Latin America & Caribbean",
    "Brazil":                            "Latin America & Caribbean",
    "British Virgin Islands":            "Latin America & Caribbean",
    "Cayman Islands":                    "Latin America & Caribbean",
    "Chile":                             "Latin America & Caribbean",
    "Colombia":                          "Latin America & Caribbean",
    "Costa Rica":                        "Latin America & Caribbean",
    "Dominica":                          "Latin America & Caribbean",
    "Dominican Republic":                "Latin America & Caribbean",
    "Ecuador":                           "Latin America & Caribbean",
    "El Salvador":                       "Latin America & Caribbean",
    "French Guiana":                     "Latin America & Caribbean",
    "Grenada":                           "Latin America & Caribbean",
    "Guadeloupe":                        "Latin America & Caribbean",
    "Guyana":                            "Latin America & Caribbean",
    "Haiti":                             "Latin America & Caribbean",
    "Honduras":                          "Latin America & Caribbean",
    "Jamaica":                           "Latin America & Caribbean",
    "Martinique":                        "Latin America & Caribbean",
    "Mexico":                            "Latin America & Caribbean",
    "Montserrat":                        "Latin America & Caribbean",
    "Nicaragua":                         "Latin America & Caribbean",
    "Panama":                            "Latin America & Caribbean",
    "Paraguay":                          "Latin America & Caribbean",
    "Peru":                              "Latin America & Caribbean",
    "Puerto Rico":                       "Latin America & Caribbean",
    "Saint Kitts and Nevis":             "Latin America & Caribbean",
    "Saint Lucia":                       "Latin America & Caribbean",
    "Saint Pierre and Miquelon":         "Latin America & Caribbean",
    "Saint Vincent and the Grenadines":  "Latin America & Caribbean",
    "Turks and Caicos Islands":          "Latin America & Caribbean",
    "United States Virgin Islands":      "Latin America & Caribbean",
    "Uruguay":                           "Latin America & Caribbean",
    # Sub-Saharan Africa
    "Benin":                             "Sub-Saharan Africa",
    "Burkina Faso":                      "Sub-Saharan Africa",
    "Cape Verde":                        "Sub-Saharan Africa",
    "Central African Republic":          "Sub-Saharan Africa",
    "Chad":                              "Sub-Saharan Africa",
    "Comoros":                           "Sub-Saharan Africa",
    "Djibouti":                          "Sub-Saharan Africa",
    "Equatorial Guinea":                 "Sub-Saharan Africa",
    "Eritrea":                           "Sub-Saharan Africa",
    "Ethiopia":                          "Sub-Saharan Africa",
    "Gambia":                            "Sub-Saharan Africa",
    "Guinea":                            "Sub-Saharan Africa",
    "Guinea-Bissau":                     "Sub-Saharan Africa",
    "Kenya":                             "Sub-Saharan Africa",
    "Lesotho":                           "Sub-Saharan Africa",
    "Liberia":                           "Sub-Saharan Africa",
    "Malawi":                            "Sub-Saharan Africa",
    "Mali":                              "Sub-Saharan Africa",
    "Mauritania":                        "Sub-Saharan Africa",
    "Mauritius":                         "Sub-Saharan Africa",
    "Namibia":                           "Sub-Saharan Africa",
    "Nigeria":                           "Sub-Saharan Africa",
    "Reunion":                           "Sub-Saharan Africa",
    "Rwanda":                            "Sub-Saharan Africa",
    "Saint Helena":                      "Sub-Saharan Africa",
    "Sao Tome and Principe":             "Sub-Saharan Africa",
    "Senegal":                           "Sub-Saharan Africa",
    "Seychelles":                        "Sub-Saharan Africa",
    "Sierra Leone":                      "Sub-Saharan Africa",
    "Somalia":                           "Sub-Saharan Africa",
    "South Africa":                      "Sub-Saharan Africa",
    "South Sudan":                       "Sub-Saharan Africa",
    "Togo":                              "Sub-Saharan Africa",
    "Uganda":                            "Sub-Saharan Africa",
    # Middle East & North Africa
    "Egypt":                             "Middle East & North Africa",
    "Iran":                              "Middle East & North Africa",
    "Jordan":                            "Middle East & North Africa",
    "Kuwait":                            "Middle East & North Africa",
    "Lebanon":                           "Middle East & North Africa",
    "Morocco":                           "Middle East & North Africa",
    "Oman":                              "Middle East & North Africa",
    "Palestine":                         "Middle East & North Africa",
    "Qatar":                             "Middle East & North Africa",
    "Saudi Arabia":                      "Middle East & North Africa",
    "Sudan":                             "Middle East & North Africa",
    "Tunisia":                           "Middle East & North Africa",
    "United Arab Emirates":              "Middle East & North Africa",
    "Western Sahara":                    "Middle East & North Africa",
    "Yemen":                             "Middle East & North Africa",
    # South & Southeast Asia
    "Bangladesh":                        "South & Southeast Asia",
    "Bhutan":                            "South & Southeast Asia",
    "Cambodia":                          "South & Southeast Asia",
    "East Timor":                        "South & Southeast Asia",
    "India":                             "South & Southeast Asia",
    "Indonesia":                         "South & Southeast Asia",
    "Laos":                              "South & Southeast Asia",
    "Malaysia":                          "South & Southeast Asia",
    "Maldives":                          "South & Southeast Asia",
    "Myanmar":                           "South & Southeast Asia",
    "Pakistan":                          "South & Southeast Asia",
    "Philippines":                       "South & Southeast Asia",
    "Singapore":                         "South & Southeast Asia",
    "Sri Lanka":                         "South & Southeast Asia",
    "Thailand":                          "South & Southeast Asia",
    "Vietnam":                           "South & Southeast Asia",
    # East Asia & Pacific
    "American Samoa":                    "East Asia & Pacific",
    "Australia":                         "East Asia & Pacific",
    "China":                             "East Asia & Pacific",
    "Cook Islands":                      "East Asia & Pacific",
    "Fiji":                              "East Asia & Pacific",
    "French Polynesia":                  "East Asia & Pacific",
    "Guam":                              "East Asia & Pacific",
    "Hong Kong":                         "East Asia & Pacific",
    "Japan":                             "East Asia & Pacific",
    "Kiribati":                          "East Asia & Pacific",
    "Macao":                             "East Asia & Pacific",
    "Mongolia":                          "East Asia & Pacific",
    "Nauru":                             "East Asia & Pacific",
    "New Caledonia":                     "East Asia & Pacific",
    "New Zealand":                       "East Asia & Pacific",
    "Niue":                              "East Asia & Pacific",
    "Papua New Guinea":                  "East Asia & Pacific",
    "Samoa":                             "East Asia & Pacific",
    "Solomon Islands":                   "East Asia & Pacific",
    "South Korea":                       "East Asia & Pacific",
    "Taiwan":                            "East Asia & Pacific",
    "Tonga":                             "East Asia & Pacific",
    "Vanuatu":                           "East Asia & Pacific",
    # Eastern Europe & Central Asia
    "Armenia":                           "Eastern Europe & Central Asia",
    "Azerbaijan":                        "Eastern Europe & Central Asia",
    "Belarus":                           "Eastern Europe & Central Asia",
    "Bosnia and Herzegovina":            "Eastern Europe & Central Asia",
    "Bulgaria":                          "Eastern Europe & Central Asia",
    "Croatia":                           "Eastern Europe & Central Asia",
    "Czechia":                           "Eastern Europe & Central Asia",
    "Estonia":                           "Eastern Europe & Central Asia",
    "Georgia":                           "Eastern Europe & Central Asia",
    "Kazakhstan":                        "Eastern Europe & Central Asia",
    "Kosovo":                            "Eastern Europe & Central Asia",
    "Kyrgyzstan":                        "Eastern Europe & Central Asia",
    "Latvia":                            "Eastern Europe & Central Asia",
    "Lithuania":                         "Eastern Europe & Central Asia",
    "Moldova":                           "Eastern Europe & Central Asia",
    "Montenegro":                        "Eastern Europe & Central Asia",
    "North Macedonia":                   "Eastern Europe & Central Asia",
    "Poland":                            "Eastern Europe & Central Asia",
    "Romania":                           "Eastern Europe & Central Asia",
    "Russia":                            "Eastern Europe & Central Asia",
    "Serbia":                            "Eastern Europe & Central Asia",
    "Slovakia":                          "Eastern Europe & Central Asia",
    "Slovenia":                          "Eastern Europe & Central Asia",
    "Tajikistan":                        "Eastern Europe & Central Asia",
    "Turkey":                            "Eastern Europe & Central Asia",
    "Turkmenistan":                      "Eastern Europe & Central Asia",
    "Ukraine":                           "Eastern Europe & Central Asia",
    "Uzbekistan":                        "Eastern Europe & Central Asia",
    # Western Europe & North America
    "Austria":                           "Western Europe & North America",
    "Belgium":                           "Western Europe & North America",
    "Canada":                            "Western Europe & North America",
    "Cyprus":                            "Western Europe & North America",
    "Denmark":                           "Western Europe & North America",
    "Falkland Islands":                  "Western Europe & North America",
    "Faroe Islands":                     "Western Europe & North America",
    "Finland":                           "Western Europe & North America",
    "France":                            "Western Europe & North America",
    "Germany":                           "Western Europe & North America",
    "Gibraltar":                         "Western Europe & North America",
    "Greece":                            "Western Europe & North America",
    "Hungary":                           "Western Europe & North America",
    "Iceland":                           "Western Europe & North America",
    "Ireland":                           "Western Europe & North America",
    "Italy":                             "Western Europe & North America",
    "Luxembourg":                        "Western Europe & North America",
    "Malta":                             "Western Europe & North America",
    "Netherlands":                       "Western Europe & North America",
    "Norway":                            "Western Europe & North America",
    "Portugal":                          "Western Europe & North America",
    "Spain":                             "Western Europe & North America",
    "Sweden":                            "Western Europe & North America",
    "Switzerland":                       "Western Europe & North America",
    "United Kingdom":                    "Western Europe & North America",
    "United States":                     "Western Europe & North America",
}

print(f"Missing BEFORE regional GDP imputation: {df_clean['gdp'].isna().sum()}")

# Step 1: assign region to every row
df_clean["_region"] = df_clean["country"].map(REGION_MAP)

# Step 2: GDP per capita for countries that have both GDP and population
df_clean["_gdp_pc"] = df_clean["gdp"] / df_clean["population"]

# Step 3: median GDP per capita per (region, year) from countries WITH data
regional_median = (
    df_clean[df_clean["gdp"].notna() & df_clean["_region"].notna()]
    .groupby(["_region", "year"])["_gdp_pc"]
    .median()
    .reset_index()
    .rename(columns={"_gdp_pc": "_regional_median_gdp_pc"})
)

# Step 4: merge and impute
df_clean = df_clean.merge(regional_median, on=["_region", "year"], how="left")

mask = (
    df_clean["gdp"].isna()
    & df_clean["_regional_median_gdp_pc"].notna()
    & df_clean["population"].notna()
)
df_clean.loc[mask, "gdp"] = (
    df_clean.loc[mask, "_regional_median_gdp_pc"] * df_clean.loc[mask, "population"]
)

filled = mask.sum()
remaining = df_clean["gdp"].isna().sum()
print(f"  Filled {filled} GDP cells using regional median GDP per capita")
print(f"Missing AFTER regional GDP imputation: {remaining}")

# Per-region summary
print("\n── Fill summary by region ──")
for region in sorted(set(REGION_MAP.values())):
    countries = [c for c, r in REGION_MAP.items() if r == region]
    n_missing = df_clean[df_clean["country"].isin(countries)]["gdp"].isna().sum()
    print(f"  {region:<35} remaining NaN: {n_missing}")

# Drop helper columns
df_clean.drop(columns=["_region", "_gdp_pc", "_regional_median_gdp_pc"], inplace=True)
```
