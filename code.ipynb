```python
import pandas as pd
```

```python
df = pd.read_csv('./owid-energy-data.csv')
df.head()
```

```python
columns_needed = [
    "country",
    "year",
    "population",
    "iso_code",
    "gdp",
    "biofuel_electricity",
    "coal_electricity",
    "electricity_demand",
    "electricity_generation",
    "fossil_electricity",
    "gas_electricity",
    "greenhouse_gas_emissions",
    "hydro_electricity",
    "nuclear_electricity",
    "oil_electricity",
    "renewables_electricity",
    "solar_electricity",
    "wind_electricity",
]

df = df[columns_needed]
df.head()
```

```python
# Find missing values
missing_values = df.isnull().sum()


for column in df.columns:
    print(f"{column}: {missing_values[column]/len(df)*100:.2f}")
```

```python
# 1. Number of years of data per country
years_per_country = df.groupby("country").size()

# 2. Count missing values per country for each variable
missing_counts = (
    df
    .groupby("country")
    .apply(lambda x: x.isna().sum())
)

# 3. Convert counts to percentages (normalized by years available)
missing_pct = missing_counts.div(years_per_country, axis=0) * 100

# 4. Inspect results
missing_pct.tail()
```

```python
df = df[df["year"] >= 1985].copy()

print("Dataset restricted to years >= 1985.")
print(f"Year range after filtering: {df['year'].min()} - {df['year'].max()}")
```

```python
# Columns that must meet the data-quality threshold
critical_cols = [
    "iso_code",
    "population",
    "electricity_generation",
    "electricity_demand"
]

# Identify countries that violate the 20% missing-data rule
countries_to_drop = missing_pct[
    (missing_pct[critical_cols] > 50).any(axis=1)
].index

# Drop those countries from the main dataframe
df_clean = df[~df["country"].isin(countries_to_drop)].copy()

print(f"Dropped {len(countries_to_drop)} countries due to excessive missing data.")
print(f"Remaining countries: {df_clean['country'].nunique()}")
```

solar and wind need to be handled differently from the other columns.
looking at the raw data, every pre-2000 non-null value for both columns is
already 0.0 - so the NaNs before 2000 aren't really missing, they just mean
the country had no capacity yet. same logic applies to post-2000 leading NaNs
(e.g. Bosnia had no wind farms until 2011, so those early 2000s rows should
also be zero, not interpolated). for 2023/2024 it's just a reporting lag so
we carry the last known value forward.

```python

renewable_cols = ["solar_electricity", "wind_electricity"]

def impute_solar_wind(df, col):
    result = df[col].copy()

    for country, group in df.groupby("country"):
        idx = group.index
        years = group["year"]
        vals = group[col]

        # anything before 2000 is a true zero
        pre2000 = idx[years < 2000]
        result.loc[pre2000] = result.loc[pre2000].fillna(0.0)

        post2000_idx = idx[years >= 2000]
        post2000_vals = vals.loc[post2000_idx]

        # if there's no post-2000 data at all just zero fill and move on
        if post2000_vals.isna().all():
            result.loc[post2000_idx] = 0.0
            continue

        first_valid = post2000_vals.first_valid_index()
        last_valid = post2000_vals.last_valid_index()

        # leading NaNs post-2000 = country hadn't built capacity yet, so zero
        leading = post2000_idx[post2000_idx < first_valid]
        result.loc[leading] = result.loc[leading].fillna(0.0)

        # trailing NaNs = 2023/2024 reporting lag, carry last known value
        trailing = post2000_idx[post2000_idx > last_valid]
        if len(trailing) > 0:
            result.loc[trailing] = result.loc[trailing].fillna(post2000_vals.loc[last_valid])

    return result


for col in renewable_cols:
    before = df_clean[col].isna().sum()
    df_clean[col] = impute_solar_wind(df_clean, col)
    after = df_clean[col].isna().sum()
    print(f"{col}: {before} missing -> {after} missing (filled {before - after})")


# quick checks
print()
for col in renewable_cols:
    print(f"{col} negatives: {(df_clean[col] < 0).sum()}")
    pre2000 = df_clean.loc[df_clean["year"] < 2000, col]
    print(f"{col} pre-2000 remaining NaN: {pre2000.isna().sum()}")

# spot check a few countries that had leading post-2000 NaNs
spot_check = {
    "Bosnia and Herzegovina": ("wind_electricity", 2000, 2010),
    "Malta": ("wind_electricity", 2000, 2011),
    "Martinique": ("solar_electricity", 2000, 2005),
}
print()
for country, (col, y1, y2) in spot_check.items():
    vals = df_clean[
        (df_clean["country"] == country) & df_clean["year"].between(y1, y2)
    ][["year", col]]
    print(f"{country} {col} {y1}-{y2}:")
    print(vals.to_string(index=False), "\n")

# check trailing fill worked
print("Uganda solar 2021-2024:")
print(df_clean[(df_clean["country"] == "Uganda") & (df_clean["year"] >= 2021)][["year", "solar_electricity"]].to_string(index=False))
```
